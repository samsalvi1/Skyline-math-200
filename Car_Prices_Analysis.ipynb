```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
link = 'https://raw.githubusercontent.com/samsalvi1/Skyline-math-200/main/Cars'
```

```python
Cars = pd.read_csv(link)
Cars
```

```python
Cars.corr()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
Y = Cars['CAR_PRICE']
X = Cars['Car_Age']
X = sm.add_constant(X)
reg_Equation= sm.OLS(Y,X ).fit()
print(reg_Equation.summary())
```

```python
sns.regplot(x = 'MILEAGE', y = 'CAR_PRICE', data = Cars)
```
